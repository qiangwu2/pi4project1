```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 



```

```python
data = pd.read_csv('./data_wine.csv')
data.head()
```

```python
labels = list(data.columns)
labels 
```

```python
for columns in data.columns: 
    data = data.rename(columns={columns:columns.strip()})
```

```python
data.loc[:6,: ]
```

```python
def zscore(value):
    result= (value-value.mean())/value.std()
    return result 
```

```python
data_scaled= zscore(data.loc[ :,'Alcohol':'Proline'])
data_scaled.head()
```

```python
data_scaled.insert(0,'Category',data['Category'])
```

```python
data_scaled.head()
```

```python
conMatrix = data_scaled.loc[:,'Alcohol':'Proline'].cov()
print(conMatrix)
```

```python
from biokit.viz import corrplot 
ax=corrplot.Corrplot(data_scaled.loc[:,'Alcohol':'Proline'].cov())
```

```python
ax.plot()
plt.show()
```

```python
from numpy.linalg import eig
eigV = eig(conMatrix)
eigVal= eigV[0]
eigVec = eigV[1]
```

```python
eigV
```

```python
infor =0.8
sortedValue = sorted(eigVal,reverse=True)
sortedIndex= sorted(range(len(eigVal)),key=lambda k:eigVal[k],reverse=True)
```

```python
sortedValue
sortedIndex
```

```python
fig,ax = plt.subplots()
ax.plot(range(1,len(sortedValue)+1),np.cumsum(sortedValue)/np.sum(sortedValue),'mo--')
ax.plot(range(1,len(sortedValue)+1),np.ones(len(sortedValue))*infor,'b-')
plt.xlabel("Numer of principla conpoments")
plt.ylabel("percentage of information")
plt.show()
```

```python
eigIDs =[]
for i in range(0,len(sortedValue)):
    if sum(sortedValue[0:i+1])/sum(sortedValue) > infor:
        eigIDs = sortedIndex[0:i+1]
        break 
    
eigVal_pca = eigVal[eigIDs]
print(eigVal_pca)
eigVec_pca = eigVec[ :,eigIDs]
print(eigVec_pca)
```

```python
data_array = data_scaled.loc[:,'Alcohol':'Proline'].as_matrix()
data_pca = data_array.dot(eigVec_pca)
plt.plot(data_pca[:,0],data_pca[:,1],'bo')
plt.show()
```
